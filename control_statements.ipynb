```python
#==================================
#========== IF ELIF ELSE ==========
#==================================
# if-else
grade1 = 87
if grade1 >= 90:
    print("You get an A")
else:
    print("You did not get an A")

# if-elif-else
if grade1 >= 90:
    print("You get an A")
elif grade1 >= 80:
    print("You get a B")
else:
    print("You got a C or worse, loser.")

# else is optional

```

```python
#==================================
#============= WHILE ==============
#==================================
number1 = 11
while number1 < 20:
    print(number1, "< 20 = True, Incrementing.")
    number1 += 1
print(number1, "< 20 is no longer true")
```

```python
#==================================
#============= FOR ================
#==================================
some_range = [2, -3, 0, 17, 9] # note this is an iterable
total = 0
for value in some_range:
    total += value
print("Total: ", total)

# counter defaults to 0
# range(x) is a function
for counter in range(10):
    print(counter, end=' ')
```

```python
#==================================
#======== BREAK/CONTINUE ==========
#==================================
for num in range(10):
    if num == 5:
        break
    print(num, end=' ')
print("\n")

for num in range(10):
    if num == 6:
        continue
    print(num, end=' ')
```
