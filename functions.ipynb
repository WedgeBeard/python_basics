```python
#==================================
#============ FUNCTIONS ===========
#==================================
def square(number):
    """Calculates the square of a number. That's it.""" # doc string for IDE to show when using this function
    return number ** 2

print(square(8))

# IPython's help mechanism: <thing I want info about>?
square?
```

```python
#==================================
#======= DEFAULT PARAMETERS =======
#==================================
def say_my_name(first_name="Pagan", last_name="Bob"):
    print(f"My name is {first_name} {last_name}")

say_my_name()
say_my_name("Jim", "Dexter")
say_my_name("Sinky")

# keyword arguments - order doesn't matter since arguments are identified by name
say_my_name(last_name="Strickland", first_name="Maniel")
say_my_name(last_name="Strickland")
```

```python
#==================================
#======= ARBITRARY ARG LIST =======
#==================================
def average(*args):
    return sum(args) / len(args) 
print(average(10,100))
print(average(3, 65, 22, 35, 101))

# unpack an iterable passed in as a variable
number_list = [5, 23, 11, 194, 24, 54]
print(average(*number_list))
```
