```python
#==================================
#============= RANDOM =============
#==================================
import random

random.seed(None) # adding this so I don't have to Restart and clear cache every time
print("Rolling 10 dice:")
for roll in range(10):
    print(random.randrange(1,7), end=' ')

# Use a seed for predictable randomness
random.seed(444)
print("\nRolling 10 dice with seed:")
for roll in range(10):
    print(random.randrange(1,7), end=' ')
print("\nRolling 10 dice again with seed:")
for roll in range(10):
    print(random.randrange(1,7), end=' ')
```

```python
die1 = random.randrange(1,7)
die2 = random.randrange(1,7)
die_tuple = (die1, die2)
sum_of_dice = sum(die_tuple)
print(sum_of_dice)

if sum_of_dice in (7, 11): # same as -> if (sum_of_dice == 7) or (sum_of_dice == 11):
    print("You win!")
```
