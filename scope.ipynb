```python
#==================================
#============= SCOPE ==============
#==================================

# global variable
my_var = 4

def access_global_var():
    print("From access_global_var, my_var = ", my_var)

access_global_var()

def change_global_var():
    my_var = 14 # actually shadow variable for the global version
    print("From change_global_var, my_var = ", my_var)

change_global_var()

print(my_var)

def change_global_var_really():
    global my_var # declare it in the function with "global"
    my_var = 16 # then assign
    print("From change_global_var_really, my_var = ", my_var)

change_global_var_really()

print(my_var)
```
