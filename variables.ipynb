```python
#==================================
#======== DATA COLLECTIONS ========
#==================================

# use [] for lists
# can contain homogeneous or heterogeneous types (same type, or mixed types)
# MUTABLE
my_list = ["Age", 47.56, 'A', "bruh", 67]
print(type(my_list))
print("element 2: ", my_list[2])
print("element -1: ", my_list[-1])
print("element -2: ", my_list[-2])


# use {} for sets
my_set = {"first", 1, "second", 2}
print(type(my_set))

# use () for tuples
my_tuple = ("Words", 12, 123.33)
print(type(my_tuple))

```
