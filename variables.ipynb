```python
#==================================
#======== DATA COLLECTIONS ========
#==================================

# use [] for lists
# can contain homogeneous or heterogeneous types (same type, or mixed types)
# MUTABLE
my_list = ["Age", 47.56, 'A', "bruh", 67]
print(type(my_list))
print("element 2: ", my_list[2])
print("element -1: ", my_list[-1])
print("element -2: ", my_list[-2])


# use {} for sets
my_set = {"first", 1, "second", 2}
print(type(my_set))

# use () for tuples
# IMMUTABLE
my_tuple = ("Words", 12, 123.33) # called "packing a tuple", enclosing with ()
print(type(my_tuple))

# unpacking a tuple
the_word, the_number, the_float = my_tuple
print(f"The word: {the_word}, the number: {the_number}, the float: {the_float}")

```
