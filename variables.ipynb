```python
#==================================
#======== DATA COLLECTIONS ========
#==================================

# use [] for lists
# can contain homogeneous or heterogeneous types (same type, or mixed types)
# MUTABLE
my_list = ["Age", 47.56, 'A', "bruh", 67]
print(type(my_list))
print("element 2: ", my_list[2])
print("element -1: ", my_list[-1])
print("element -2: ", my_list[-2])


# use {} for sets
my_set = {"first", 1, "second", 2}
print(type(my_set))

# use () for tuples
# IMMUTABLE
my_tuple = ("Words", 12, 123.33) # called "packing a tuple", enclosing with ()
print(type(my_tuple))

# unpacking a tuple
the_word, the_number, the_float = my_tuple
print(f"The word: {the_word}, the number: {the_number}, the float: {the_float}")

```

```python
#==================================
#======== MUTABLE/IMMUTABLE =======
#==================================

# check mutability
# if a number is returned, it's immutable
# if a TypeError returns, it's mutable

# IMMUTABLE
my_int = 12
my_float = 12.22
my_char = 'a'
my_string = "Bob's a sucka!"
my_bool = True
my_tuple = ("stuff", 1, 12.3)

# MUTABLE
my_list = [12, 23, 5454, 2]
my_set = {1, "this", 'a', 24.2}
my_dictionary = {1: "first", 2: "second"}

print("my_int: ", hash(my_int))
print("my_float: ", hash(my_float))
print("my_char: ", hash(my_char))
print("my_string: ", hash(my_string))
print("my_bool: ", hash(my_bool))
print("my_tuple: ", hash(my_tuple))
# print("my_list: ", hash(my_list)) # mutable
# print("my_set: ", hash(my_set)) # mutable
# print("my_dictionary: ", hash(my_dictionary)) # mutable
```
